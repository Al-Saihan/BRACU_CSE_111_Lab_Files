```python
# Class ----------------------------------------------------------------------------------
class Student:
    def __init__(self, name, gpa, creds=9, dep="CSE"):
        self.name = name
        self.cgpa = gpa
        self.creds = creds
        self.dep = dep

        if self.cgpa >= 3.5 and self.creds > 10:
            if self.cgpa >= 3.7:
                self.scholar = "Merit-based scholarship"
            elif 3.5 <= self.cgpa < 3.7:
                self.scholar = "Need-based scholarship"
        else:
            self.scholar = "No scholarship"

    def checkScholarshipEligibility(self):
        if self.scholar == "Merit-based scholarship":
            print(f"{self.name} is eligible for Merit-Based scholarship.")
        elif self.scholar == "Need-based scholarship":
            print(f"{self.name} is eligible for Need-based scholarship.")
        else:
            print(f"{self.name} is not eligible for scholarship.")

    def showDetails(self):
        print(f"Name: {self.name}")
        print(f"Department: {self.dep}")
        print(f"CGPA: {self.cgpa}")
        print(f"Number of credits: {self.creds}")
        print(f"Scholarship Status: {self.scholar}")


# Driver Code ----------------------------------------------------------------------------------
print("1--------------------------")
std1 = Student("Alif", 3.99, 12)
print("2--------------------------")
std1.checkScholarshipEligibility()
print("3--------------------------")
std1.showDetails()
print("4--------------------------")
std2 = Student("Mim", 3.4)
std3 = Student("Henry", 3.5, 15, "BBA")
print("5--------------------------")
std2.checkScholarshipEligibility()
print("6--------------------------")
std3.checkScholarshipEligibility()
print("7--------------------------")
std2.showDetails()
print("8--------------------------")
std3.showDetails()
print("9--------------------------")
std4 = Student("Bob", 4.0, 6, "CSE")
print("10--------------------------")
std4.checkScholarshipEligibility()
print("11--------------------------")
std4.showDetails()
```

```python
# Class ----------------------------------------------------------------------------------
class Foodie:
    def __init__(self, name):
        self.name = name
        self.items = []
        self.spent = 0

    def show_orders(self):
        print(f"{self.name} has {len(self.items)} item(s) in the cart.")
        print(f"Items: {self.items}")
        return f"Total Spent: {self.spent}"

    def order(self, *list0):
        for i in list0:
            i = i.split("-")
            item, qty = i[0], int(i[1])
            price = menu[item] * qty
            print(
                f"Ordered - {item}, quantity - {qty}, price (per Unit) - {menu[item]}."
            )
            print(f"Total Price - {price}")

            self.items.append(item)
            self.spent += price

    def pay_tips(self, tip=None):
        if tip == None:
            print("No tips to the waiter")
        else:
            print(f"Gives {tip}/- tips to the waiter.")
            self.spent += tip


# Driver Code ----------------------------------------------------------------------------------
menu = {
    "Chicken Lollipop": 15,
    "Beef Nugget": 20,
    "Americano": 180,
    "Red Velvet": 150,
    "Prawn Tempura": 80,
    "Saute Veg": 200,
}

f1 = Foodie("Frodo")
print(f1.show_orders())
print("1----------------------")
f1.order("Chicken Lollipop-3", "Beef Nugget-6", "Americano-1")
print("2----------------------")
print(f1.show_orders())
print("3----------------------")
f1.order("Red Velvet-1")
print("4----------------------")
f1.pay_tips(20)
print("5----------------------")
print(f1.show_orders())
f2 = Foodie("Bilbo")
print("6----------------------")
f2.order("Prawn Tempura-6", "Saute Veg-1")
print("7----------------------")
f2.pay_tips()
print("8----------------------")
print(f2.show_orders())
```

```python
# Class ----------------------------------------------------------------------------------
class Department:
    def __init__(self, dep="ChE Department", sec=5):
        self.dep = dep
        self.sec = sec
        self.studs = 0
        print(f"The {dep} has {sec} sections.")

    def add_students(self, *students):
        if len(students) == self.sec:
            for i in students:
                self.studs += i
            print(
                f"The {self.dep} has an average of {self.studs / self.sec} students in each section."
            )
        else:
            print(f"The {self.dep} doesn't have {len(students)} sections.")

    def merge_Department(self, *nDep):
        for obj in nDep:
            print(f"{obj.dep} is merged to {self.dep}.")
            self.studs += obj.studs

        return f"Now the {self.dep} has an average of {self.studs / self.sec}."


# Driver Code ----------------------------------------------------------------------------------
d1 = Department()
print("1-----------------------------")
d2 = Department("MME Department")
print("2-----------------------------")
d3 = Department("NCE Department", 8)
print("3-----------------------------")
d1.add_students(12, 23, 12, 34, 21)
print("4-----------------------------")
d2.add_students(40, 30, 21)
print("5-----------------------------")
d3.add_students(12, 34, 41, 17, 30, 22, 32, 51)
print("6-----------------------------")
mega = Department("Engineering Department", 10)
print("7-----------------------------")
mega.add_students(21, 30, 40, 36, 10, 32, 27, 51, 45, 15)
print("8-----------------------------")
print(mega.merge_Department(d1, d2))
print("9-----------------------------")
print(mega.merge_Department(d3))
```

```python
# Class ----------------------------------------------------------------------------------
class Shopidify:
    def __init__(self, name="Guest"):
        self.cart = {}
        self.history = []
        self.name = name
        if name == "Guest":
            print("Welcome to Shopidify")
        else:
            print(f"Welcome {name} to Shopidify")

    def add_to_cart(self, *items):
        if isinstance(items[0], list):
            for i in range(0, len(items[0]), 2):
                if items[0][1] not in self.cart:
                    self.cart[items[0][i]] = items[0][i + 1]
                else:
                    self.cart[items[0][i]] += items[0][i + 1]
        else:
            if len(items) == 1:
                if items[0] not in self.cart:
                    self.cart[items[0]] = 1
                else:
                    self.cart[items[0]] += 1
            else:
                for i in range(0, len(items), 2):
                    if items[i] not in self.cart:
                        self.cart[items[i]] = items[i + 1]
                    else:
                        self.cart[items[i]] += items[i + 1]

    def display_cart(self):
        print(f"Items in the cart for {self.name}:")
        for item, qty in self.cart.items():
            print(f"- {item}: {qty}x")

    def checkout(self):
        print(f"Checkout completed for {self.name}")
        self.history.append(self.cart.copy())
        self.cart = {}

    def display_history(self):
        print(f"Purchase History for {self.name}:")
        for trans in range(len(self.history)):
            print(f"Transaction {trans+1}")
            for item, qty in self.history[trans].items():
                print(f"- {item}: {qty}x")


# Driver Code ----------------------------------------------------------------------------------
guest_account = Shopidify()
print("1xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account = Shopidify("John")
print("2xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.add_to_cart("Air Jordan", 2)
guest_account.add_to_cart("Luffy Action Figure")
guest_account.display_cart()
print("3xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.add_to_cart(
    ["Chocolate Chip Cookies", 3, "Goku Action Figure", 2, "Dumbbells-5kg", 2]
)
john_account.display_cart()
print("4xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.add_to_cart("Air Jordan")
guest_account.display_cart()
print("5xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.checkout()
print("6xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.display_history()
print("7xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.checkout()
print("8xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.display_history()
print("9xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
```

```python
# Class ----------------------------------------------------------------------------------
class Author:
    def __init__(self, name=None):
        self.name = name
        self.books = {}
        self.total = 0

    def addBook(self, bookName, genre):
        if self.name == None:
            print("A book can not be added without author name")
        else:
            if genre not in self.books:
                self.books[genre] = bookName
            else:
                self.books[genre] += f", {bookName}"
            self.total += 1

    def setName(self, name):
        self.name = name

    def printDetail(self):
        print(f"Number of Book(s): {self.total}")
        print(f"Author Name: {self.name}")
        for key in self.books:
            print(f"{key}: {self.books[key]}")


# Driver Code ----------------------------------------------------------------------------------
a1 = Author()
print("==========================")
a1.addBook("Ice", "Science Fiction")
print("==========================")
a1.setName("Anna Kavan")
a1.addBook("Ice", "Science Fiction")
a1.printDetail()
print("==========================")
a2 = Author("Humayun Ahmed")
a2.addBook("Onnobhubon", "Science Fiction")
a2.addBook("Megher Upor Bari", "Horror")
print("==========================")
a2.printDetail()
a2.addBook("Ireena", "Science Fiction")
print("==========================")
a2.printDetail()
print("==========================")
```
