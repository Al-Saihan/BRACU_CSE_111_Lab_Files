```python
#Task 1
class CellPackage:
    def __init__(self, price , data, talktime, msg, cashback ,validity):
        self.price = price
        self.data = int(data.split()[0]) * 1024 # GB --> MB
        self.talk_time = talktime
        self.messages = msg
        self.validity = validity
        discount = float(cashback[-len(cashback) : -1 : 1]) / 100
        self.cashback = int(self.price * discount)
    
    def check(self):
        if self.data != 0:
            print(f"Data = {self.data} MB")
        if self.talk_time != 0:
            print(f"Talktime = {self.talk_time} Minutes")
        if self.messages != 0:
            print(f"SMS/MMS = {self.messages}")
        print(f"Validity = {self.validity} Days")
        print(f"--> Price = {self.price} tk")
        if self.cashback != 0:
            print(f"Buy now to get {self.cashback} tk cashback")

pkg = CellPackage(150, '6 GB', 99, 20, '7%', 7)
print('============= Package 1 =============')
pkg.check()

pkg2 = CellPackage(700, '35 GB', 700, 0, '10%', 30)
print('============= Package 2 =============')
pkg2.check()

pkg4 = CellPackage(120, '0 GB', 190, 0, '0%', 10)
print('============= Package 3 =============')
pkg4.check()
```

```python
#Task 4
class abcTech:
    def __init__(self, name , des, dep):
        self.name = name
        self.designation = des
        self.depertment = dep
        self.skills = []
        self.frameworks = []
        print(f"Welcome to abcTech, {self.name}!")
    
    def addProgrammingSkills(self, skills):
        for i in skills:
            self.skills.append(i)
    
    def addFrameworks(self, frameworks):
        for i in frameworks:
            self.frameworks.append(i)
    
    def printInfo(self):
        print(f"Name: {self.name}")
        print(f"Designation: {self.designation}")
        print(f"Department: {self.depertment}")
        print("Programming skills:" , ", ".join(self.skills))
        print("Frameworks:" , ", ".join(self.frameworks))
    
    def calculateSalary(self, salary , hours):
        if hours <= 144:
            return salary
        else:
            salary += (hours - 144) * 800
            return salary
```

```python
#Task 5
class StudentDatabase:
    def __init__(self, name, id):
        self.name, self.id, self.grades = name , id, {}
    
    def calculateGPA(self, li, session):
        liPrime, gpaSum, count = [] , 0 , 0
        
        for i in li:
            x = i.split(": ")
            liPrime.append(x[0])
            gpaSum += float(x[1]) 
            count += 1
            
        self.grades[session] = {tuple(liPrime) : round((gpaSum / count) , 2)}

    def printDetails(self):
        print(f"Name: {self.name}")
        print(f"ID: {self.id}")
        
        for i in self.grades.keys():
            print(f"Courses taken in {i}")

            for j in self.grades[i].keys():
                for k in j:
                    print(k)
                    
                print(f"GPA: {self.grades[i][j]}")
```

```python
# Task 6
class Store:
    def __init__(self, bal) -> None:
        self.balance , self.items , = bal , {} 
        self.total_items  = len(self.items)
        print("New branch created!")

    def viewAllItems(self):
        if self.total_items == 0:
            print("There are no items in your inventory")
        else:
            print(f"All Items:" , ", ".join(list(self.items.keys())))
            
    
    def viewAllItemDetails(self):
        print(self.items)
    
    def add_item(self, li0):
        item, stock, buycost, sellprice = li0
        
        self.balance -= (stock * buycost)
        self.items[item] = {"stock" : stock , "buying_price" : buycost ,
                            "selling_price" : sellprice} 
                            
        print(f"Item added: {item}")
        
    def sell_item(self, item, qty):
        a = self.items[item]["stock"]
        
        if a < qty:
            print(f"Sorry! {item} us not available at your desired quantity. Currently we have: {a}") 
        else:
            a -= qty
            self.balance += self.items[item]["selling_price"] * qty

    def restock_item(self, item, qty):
        a = self.items[item]["stock"]
        a += qty
        print(f"Restocked item: {item}, Current Stock: {a}")
        self.balance -= self.items[item]["buying_price"] * qty
```
