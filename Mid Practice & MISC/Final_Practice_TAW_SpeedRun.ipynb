```python
class PlayerEarning:
    def __init__(self, name) -> None:
        self.name = name

    def calculateTotal(self, earning, goals=0):
        self.earn = earning

        if goals > 30:
            self.bonus = int(0.05 * self.earn + 10000)
        elif 0 < goals <= 30:
            self.bonus = int(0.05 * self.earn)
        else:
            self.bonus = 0

        self.total = self.earn + self.bonus

    def printDetails(self):
        print(f"Player Name: {self.name}")
        print(f"Player Season Earning without bonus: {self.earn}")
        print(f"Bonus: {self.bonus}")
        print(f"Player Season Earning After Bonus: {self.total}")


print("**********************")
player1 = PlayerEarning("Buffon")
player1.calculateTotal(250000)
player1.printDetails()
print("\n**********************")
player2 = PlayerEarning("Dybala")
player2.calculateTotal(250000, 31)
player2.printDetails()
print("\n**********************")
player3 = PlayerEarning("Cuadrado")
player3.calculateTotal(250000, 20)
player3.printDetails()
```

```python
class myList:
    def __init__(self, *args):
        self.storage = list(args)
        self.total = 0

        for i in self.storage:
            self.total += i

    def sum(self):
        print(f"Sum : {self.total}")

    def merge(self, *args):
        for i in args:
            self.storage.append(i)
            self.total += i

    def average(self):
        if len(self.storage) == 0:
            return print("Average: 0")

        print(f"Average: {self.total / len(self.storage)}")


l1 = myList(2, 3, 4, 5, 6)
# you might need a list inside your class to store the values
l1.sum()
l1.merge(4, 5, 9)
l1.sum()
l1.average()
print("-----------------------------")
l2 = myList()
l2.average()
l2.merge(1, 2, 4, 8)
l2.sum()
```

```python
class Student:
    def __init__(self, name, ID):
        self.name = name
        self.ID = ID

    def Details(self):
        return "Name: " + self.name + "\n" + "ID: " + self.ID + "\n"


# Write your code here
class CSEStudent(Student):
    def __init__(self, name, ID, sem):
        super().__init__(name, ID)
        self.sem = sem
        self.courses = {}

    def Details(self):
        return (
            "Name: "
            + self.name
            + "\n"
            + "ID: "
            + self.ID
            + "\n"
            + f"Current semester: {self.sem}"
        )

    def addCourseWithMarks(self, *args):
        for i in range(0, len(args), 2):
            if args[i + 1] >= 85:
                gpa = "4.0"
            elif args[i + 1] >= 80:
                gpa = "3.3"
            elif args[i + 1] >= 70:
                gpa = "3.0"
            elif args[i + 1] >= 65:
                gpa = "2.3"
            elif args[i + 1] >= 57:
                gpa = "2.0"
            elif args[i + 1] >= 55:
                gpa = "1.3"
            elif args[i + 1] >= 50:
                gpa = "1.0"
            else:
                gpa = "0.0"

            self.courses[args[i]] = gpa

    def showGPA(self):
        total = 0
        count = 0
        print(f"{self.name} has taken {len(self.courses)} courses.")
        for key, value in self.courses.items():
            print(f"{key}: {value}")
            total += float(value)
            count += 1
        print(f"GPA of {self.name} is : {round(total/count, 2)}")


Bob = CSEStudent("Bob", "20301018", "Fall 2020")
Carol = CSEStudent("Carol", "16301814", "Fall 2020")
Anny = CSEStudent("Anny", "18201234", "Fall 2020")
print("#########################")
print(Bob.Details())
print("#########################")
print(Carol.Details())
print("#########################")
print(Anny.Details())
print("#########################")
Bob.addCourseWithMarks("CSE111", 83.5, "CSE230", 73.0, "CSE260", 92.5)
Carol.addCourseWithMarks("CSE470", 62.5, "CSE422", 69.0, "CSE460", 76.5, "CSE461", 87.0)
Anny.addCourseWithMarks("CSE340", 45.5, "CSE321", 95.0, "CSE370", 91.0)
print("----------------------------")
Bob.showGPA()
print("----------------------------")
Carol.showGPA()
print("----------------------------")
Anny.showGPA()
```

```python
class Transport:
    total_traveller = 0

    def __init__(self, name, fare):
        self.name = name
        self.baseFare = fare

    def __str__(self):
        s = "Name: " + self.name + ", Base fare: " + str(self.baseFare)
        return s


# Write your codes here.
class Bus(Transport):
    def __init__(self, name, fare):
        super().__init__(name, fare)
        self.peeps = {}
        print(f"Base-fare of {self.name} is {self.baseFare} Taka ")
        Transport.total_traveller += 1

    def addPassengerWithBags(self, *args):
        for i in range(0, len(args), 2):
            if 3 <= args[i + 1] <= 5:
                self.peeps[args[i]] = self.baseFare + 60
            elif 5 < args[i + 1]:
                self.peeps[args[i]] = self.baseFare + 105
            else:
                self.peeps[args[i]] = self.baseFare

    def __str__(self):
        s = f"""Name: {self.name}, Base fare: {self.baseFare}
Total Passenger(s): {len(self.peeps)}
Passenger details:"""

        for key, value in self.peeps.items():
            s += f"\nName: {key}, Fare: {value}"

        return s


class Train(Transport):
    def __init__(self, name, fare):
        super().__init__(name, fare)
        self.peeps = {}
        print(f"Base-fare of {self.name} is {self.baseFare} Taka ")
        Transport.total_traveller += 1

    def addPassengerWithBags(self, *args):
        for i in range(0, len(args), 2):
            if 3 <= args[i + 1] <= 5:
                self.peeps[args[i]] = self.baseFare + 60
            elif 5 < args[i + 1]:
                self.peeps[args[i]] = self.baseFare + 105
            else:
                self.peeps[args[i]] = self.baseFare

    def __str__(self):
        s = f"""Name: {self.name}, Base fare: {self.baseFare}
Total Passenger(s): {len(self.peeps)}
Passenger details:"""

        for key, value in self.peeps.items():
            s += f"\nName: {key}, Fare: {value}"

        return s


# Do not change the following lines of code.
t1 = Bus("Volvo", 950)
print("=================================")
t1.addPassengerWithBags("David", 6, "Mike", 1, "Carol", 3)
print("=================================")
print(t1)
print("=================================")
t2 = Train("Silk City", 850)
print("=================================")
t2.addPassengerWithBags("Bob", 2, "Simon", 4)
print("=================================")
print(t2)
print("=================================")
print("Total Passengers in Transport: ", Transport.total_traveller)
```

```python
class Magical_SportsPerson:
    database = {}
    playerNo = 0

    def __init__(self, name, team, jerseyNo):
        self.name = name
        self.team = team
        self.jerseyNo = jerseyNo

    def __str__(self):
        return "Name:{}\nTeam:{}\nJerseyNo:{}".format(
            self.name, self.team, self.jerseyNo
        )


# Write your code here
class Quidditch_Player(Magical_SportsPerson):
    @classmethod
    def createPlayer(cls, name, team, jerseyNo, goals, retire="Not yet retired"):
        return cls(name, team, jerseyNo, goals, retire)

    def __init__(self, name, team, jerseyNo, goals, retire="Not yet retired"):
        super().__init__(name, team, jerseyNo)
        self.goals = goals
        self.retire = retire
        Magical_SportsPerson.playerNo += 1
        tempVar = self.name.split(" ")
        self.id = (
            f"{Magical_SportsPerson.playerNo}{tempVar[0][0]}{tempVar[1][0]}{self.goals}"
        )
        Magical_SportsPerson.database[self.id] = [
            self.name,
            self.team,
            self.jerseyNo,
            self.goals,
            self.retire,
        ]

    def __str__(self):
        print(f"Player ID: {self.id}")
        print(super().__str__())
        print(f"Goals Scored: {self.goals}")
        return f"Retirement date: {self.retire}"


print("Number of players:", Magical_SportsPerson.playerNo)
print("Player Database:", Magical_SportsPerson.database)
print("#################################")
p1 = Quidditch_Player("Harry Potter", "Gryffindor", 8, 523)
print("------Details of the player------")
print(p1)
print("#################################")
p2 = Quidditch_Player("Ronald Weasley", "Gryffindor", 13, 5)
print("------Details of the player------")
print(p2)
print("#################################")
p3 = Quidditch_Player.createPlayer(
    "George Weasley", "Gryffindor", 12, 11, "11 Magical Year, 1886"
)
print("------Details of the player------")
print(p3)
print("#################################")
print("Number of players:", Magical_SportsPerson.playerNo)
print("Player Database:", Magical_SportsPerson.database)
```

```python
class User:
    def __init__(self, name, destination, lookingFor="Single"):
        self.name = name
        self.destination = destination
        self.lookingFor = lookingFor
        self.rideStatus = False

    def status(self):
        if self.rideStatus == False:
            print(f"Status: {self.name} is looking for a {self.lookingFor} ride!")
        else:
            print(f"Status: {self.name} boarded in car {self.rideStatus}")


class Uber:
    def __init__(self, number, type, *routes) -> None:
        self.number = number
        self.type = type
        self.routes = list(routes)

    def details(self):
        print(f"Car number: {self.number}")
        print(f"Type: {self.type}")
        print("Routes: ", end="")
        for i in self.routes:
            if len(self.routes) == 1:
                print(f"Routes: {i}")
                break
            if i == self.routes[0]:
                print(f"{i}", end="")
            elif i == self.routes[-1]:
                print(f" --> {i}")
            else:
                print(f" --> {i}", end="")

    def pick(self, *passengers):
        for person in passengers:
            if person.lookingFor != self.type:
                print(f"{person.name} is looking for a different ride.")
            elif person.destination not in self.routes:
                print(
                    f"{person.name}'s destination is different from this car's route."
                )
            else:
                print(f"{person.name} has been picked up.")
                person.rideStatus = self.number


user1 = User("Brooks", "Banani", "Shared")
user2 = User("Jocelyn", "Uttara")
user3 = User("Robert", "Gulshan", "Shared")
user4 = User("Langdon", "Mohakhali", "Shared")


user1.status()
user2.status()
user3.status()
user4.status()
print("----------------------------------")
car1 = Uber("0K32BH", "Shared", "Mohakhali", "Banani", "Nikunja", "Uttara")
car1.details()
print("----------------------------------")
car1.pick(user1, user2, user3, user4)
print("----------------------------------")
user1.status()
user2.status()
user3.status()
user4.status()
print("----------------------------------")
car2 = Uber("5GD2BD", "Single", "Uttara")
car3 = Uber("4T12FR", "Shared", "Gulshan", "Bashundhara")
car2.details()
car3.details()
print("----------------------------------")
car2.pick(user2, user3)
print("----------------------------------")
car3.pick(user3)
print("----------------------------------")
user2.status()
user3.status()
```
