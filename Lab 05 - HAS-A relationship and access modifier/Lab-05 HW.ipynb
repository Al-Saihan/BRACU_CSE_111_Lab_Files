```python
class Student:
    def __init__(self, name, ID, cgpa) -> None:
        self.name = name
        self.__ID = ID
        self.__cgpa = cgpa

    def getID(self):
        return self.__ID

    def setId(self, ID):
        self.__ID = ID

    def getCG(self):
        return self.__cgpa


class Department:
    students = {}

    def __init__(self, dep) -> None:
        self.dep = dep

    def addStudent(self, *li):
        for obj in li:
            if obj.getID() not in Department.students:
                Department.students[obj.getID()] = obj
                print(f"Welcome to CSE department, {obj.name}")
            else:
                print(
                    "Student with the same ID already exists, Please try with another ID"
                )

    def findStudent(self, ID):
        if ID not in Department.students:
            print("Student with this ID doesn't exist, Please give a valid ID.")
        else:
            print("Student info:")
            print(f"Student Name: {Department.students[ID].name}")
            print(f"ID: {ID}")
            print(f"CGPA: {Department.students[ID].getCG()}")

    def details(self):
        print(f"Department name: {self.dep}")
        print(f"Number of student: {len(Department.students)}")
        print("Details of the students:")

        for obj in Department.students.values():
            print(f"Student name: {obj.name}, ID: {obj.getID()}, cgpa: {obj.getCG()}")


s1 = Student("Akib", 22301010, 3.29)
s2 = Student("Reza", 22101010, 3.45)
s3 = Student("Ruhan", 23101934, 4.00)
print("1==================================")
cse = Department("CSE")
cse.findStudent(22112233)
print("2==================================")
cse.addStudent(s1, s2, s3)
print("3==================================")
cse.details()
print("4==================================")
cse.findStudent(22301010)
print("5==================================")
s4 = Student("Nakib", 22301010, 3.22)
cse.addStudent(s4)
print("6==================================")
s4.setId(21201220)
cse.addStudent(s4)
print("7==================================")
cse.details()
print("8==================================")
s5 = Student("Sakib", 22201010, 2.29)
cse.addStudent(s5)
print("9==================================")
cse.details()
```

```python
class Spaceship:
    def __init__(self, name, cap) -> None:
        self.name = name
        self.capacity = cap
        self.load = 0
        self.cargoList = []

    def load_cargo(self, obj):
        if obj.getWeight() + self.load > self.capacity:
            print(
                f"Warning: Unable to load {obj.getCargo()} inside {self.name}. Exceeds capacity by {obj.getWeight() + self.load - self.capacity}"
            )
        else:
            self.load += obj.getWeight()
            self.cargoList.append(obj.getCargo())

    def display_details(self):
        print(f"Spaceship name: {self.name}")
        print(f"Capacity: {self.capacity}")
        print(f"Current Cargo Weight: {self.load}")
        print(f"Cargo: {self.cargoList}")


class Cargo:
    def __init__(self, name, weight) -> None:
        self.__name = name
        self.__weight = weight

    def getWeight(self):
        return self.__weight

    def getCargo(self):
        return self.__name


# Creating spaceships
falcon = Spaceship("Falcon", 50000)
apollo = Spaceship("Apollo", 100000)
enterprise = Spaceship("Enterprise", 220000)
print("1.===================================")
# Creating cargo
gold = Cargo("Gold", 20000)
platinum = Cargo("Platinum", 25000)
dilithium = Cargo("Dilithium", 50000)
trilithium = Cargo("Trilithium", 70000)
neutronium = Cargo("Neutronium", 80000)
print("2.===================================")
# Loading cargo onto spaceships
falcon.load_cargo(gold)
falcon.load_cargo(platinum)
falcon.display_details()
print("3.===================================")
apollo.load_cargo(gold)  # Apollo will not reach its total capacity
apollo.display_details()
print("4.===================================")
falcon.load_cargo(neutronium)  # This should exceed Falcon's capacity
print("5.===================================")
enterprise.load_cargo(dilithium)
enterprise.load_cargo(trilithium)
enterprise.load_cargo(neutronium)  # This should not exceed Enterprise's capacity
enterprise.display_details()
```

```python
class Student:
    def __init__(self, name, ID, dep) -> None:
        self.name = name
        self.ID = ID
        self.dep = dep
        self.email = None
        self.password = None
        self.login = None
        self.courses = ""
        print("Student object is created!")


class Usis:
    def __init__(self) -> None:
        print("USIS is ready to use!")

    def login(self, person):
        if person.email == None or person.password == None:
            print("Email and password need to be set.")
        else:
            self.login = True
            print("Login successful!")

    def advising(self, person, *courses):
        if self.login == None:
            print("Please login to advise courses!")
        else:
            if len(courses) == 0:
                print("You haven't selected any courses.")
            elif len(courses) > 3:
                print("You need special approval to take more than 3 courses.")
            else:
                print("Advising successful!")
                for i in courses:
                    person.courses += f"{i}, "

    def individualDetails(self, person):
        print(f"Name {person.name}")
        print(f"ID: {person.ID}")
        print(f"Department: {person.dep}")
        return f"Advised courses: {person.courses[:len(person.courses)-2]}"


rakib = Student("Rakib", 12301455, "CSE")
print("1***********************")
usis_obj = Usis()
print("2***********************")
usis_obj.login(rakib)
print("3***********************")
usis_obj.advising(rakib)
print("4***********************")
rakib.email = "rakib@hotmail.com"
rakib.password = "1234"
print("5***********************")
usis_obj.login(rakib)
print("6***********************")
usis_obj.advising(rakib)
print("7***********************")
usis_obj.advising(rakib, "CSE110", "PHY111", "MAT110", "CSE260")
print("8***********************")
usis_obj.advising(rakib, "CSE110", "PHY111", "MAT110")
print("9***********************")
print(usis_obj.individualDetails(rakib))
```

```python
class Train:
    def __init__(self, name, *places) -> None:
        self.name = name
        self.startTrain = places[0]
        self.endTrain = places[-1]
        self.places = list(places)
        self.passengers = []
        print(f"Welcome abroad on {self.name}")
        print(f"Start: {self.startTrain}")
        print(f"Destination: {self.endTrain}")

    def addPassenger(self, *passengers):
        for obj in passengers:
            print(f"{obj.name} welcome abroad")

            if obj.start == None:
                obj.start = self.startTrain
            if obj.end == None:
                obj.end = self.endTrain

            obj.fair = (self.places.index(obj.end) - self.places.index(obj.start)) * 100

            self.passengers.append(obj)

    def allPassengerDetails(self):
        for obj in self.passengers:
            print(
                f"Name: {obj.name}, Start: {obj.start}, Destination: {obj.end}, Fair: ${obj.fair}"
            )


class Passenger:
    def __init__(self, name, start=None, end=None) -> None:
        self.name = name
        self.start = start
        self.end = end
        self.fair = None


t1 = Train("T1-Express", "New York", "Manhattan", "Brooklyn", "Boston")
print("1========================")
p1 = Passenger("Naruto")
t1.addPassenger(p1)
p2 = Passenger("Sasuke", "Manhattan")
p3 = Passenger("Hinata", "Manhattan", "Brooklyn")
print("2========================")
t1.addPassenger(p2, p3)
print("3========================")
t1.allPassengerDetails()
print("4========================")
t2 = Train("Europe-Express", "London", "Paris", "Brussels", "Turkey")
print("5========================")
p4 = Passenger("Max", "London", "Brussels")
p5 = Passenger("Eleven", "Paris")
p6 = Passenger("Mike", "Brussels")
t2.addPassenger(p4, p5, p6)
print("6========================")
t2.allPassengerDetails()
```

```python
class BracuStudent:
    def __init__(self, name, home) -> None:
        self.name = name
        self.home = home
        self.passStatus = False

    def get_pass(self):
        self.passStatus = True

    def show_details(self):
        print(f"Student Name: {self.name}")
        print(f"Lives in {self.home}")
        print(f"Have Bus Pass? {self.passStatus}")


class BracuBus:
    def __init__(self, destination, seats=2):
        self.destination = destination
        self.seats = seats
        self.passengers = []

    def show_details(self):
        print(f"Bus Route: {self.destination}")
        print(f"Passengers count: {len(self.passengers)} (Max: {self.seats})")
        print(f"Passengers On Board: {self.passengers}")

    def board(self, *students):
        if len(students) == 0:
            print("No passengers!")
        else:
            for obj in students:
                if obj.passStatus == False:
                    print(f"You don't have bus pass!")
                elif obj.home != self.destination:
                    print(f"You got on the wrong bus!")
                elif len(self.passengers) >= self.seats:
                    print(f"Bus is full!")
                else:
                    print(f"{obj.name} boarded the bus.")
                    self.passengers.append(obj.name)


st1 = BracuStudent("Afif", "Mirpur")
print("1===========================")
st2 = BracuStudent("Shanto", "Motijheel")
st3 = BracuStudent("Taskin", "Mirpur")
st1.show_details()
st2.show_details()
print("2===========================")
st3.show_details()
print("3===========================")
bus1 = BracuBus("Mirpur")
bus2 = BracuBus("Azimpur", 5)
bus1.show_details()
bus2.show_details()
print("4===========================")
st2.get_pass()
st3.get_pass()
print("5===========================")
st2.show_details()
st3.show_details()
print("6===========================")
bus1.board()
print("7===========================")
bus1.board(st1, st2)
print("8===========================")
st1.get_pass()
st2.home = "Mirpur"
st1.show_details()
st2.show_details()
print("9===========================")
bus1.board(st1, st2, st3)
print("10===========================")
bus1.show_details()
```
