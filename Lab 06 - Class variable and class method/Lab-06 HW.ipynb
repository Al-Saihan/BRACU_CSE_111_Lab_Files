```python
# Write your code here for subtasks 1-6.
class Student:
    ID = 0

    def __init__(self, name, dep, age, cgpa) -> None:
        self.name = name
        self.department = dep
        self.age = age
        self.cgpa = cgpa
        Student.ID += 1
        self.ID = Student.ID

    def showDetails(self):
        print(f"ID: {self.ID}")
        print(f"Name: {self.name}")
        print(f"Department: {self.department}")
        print(f"Age: {self.age}")
        print(f"CGPA: {self.cgpa}")

    @classmethod
    def from_String(cls, code):
        code = code.split("-")
        return cls(code[0], code[1], code[2], code[3])


s1 = Student("Samin", "CSE", 21, 3.91)
s1.showDetails()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.showDetails()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.showDetails()
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.showDetails()


# Write the answer of subtask 5 here
print("\nSubTask 5: ")
print(
    """
    Class Variables are variables which are shared between all the instances
    of a Class and Instance Variables are owned seperately by each of the instances
    of the class.
    """
)

# Write the answer of subtask 6 here
print("\nSubTask 6: ")
print(
    """
    Class Methods takes a different parameter "cls" that points to the class instead of
    the object and Instance Methods takes a parameter "self" that points to the object.
    """
)
# You are not allowed to change the code above
```

```python
# Write your code here
class Passenger:
    count = 0

    def __init__(self, name):
        self.name = name
        self.fare = 450
        self.weight = 0
        Passenger.count += 1

    def set_bag_weight(self, kg):
        self.weight = kg

        if 20 < self.weight <= 50:
            self.fare = 500

        elif self.weight > 50:
            self.fare = 550

    def printDetail(self):
        print(f"Name: {self.name}")
        print(f"Bus Fare: {self.fare} Taka")


print("Total Passenger:", Passenger.count)
p1 = Passenger("Jack")
p1.set_bag_weight(90)
p2 = Passenger("Carol")
p2.set_bag_weight(10)
p3 = Passenger("Mike")
p3.set_bag_weight(25)
print("=========================")
p1.printDetail()
print("=========================")
p2.printDetail()
print("=========================")
p3.printDetail()
print("=========================")
print("Total Passenger:", Passenger.count)
```

```python
# Write your code here
class Travel:
    count = 0

    def __init__(self, source, destination):
        self.source = source
        self.destination = destination
        self.time = "1:00"
        Travel.count += 1

    def display_travel_info(self):
        print("Source:", self.source)
        print("Destination:", self.destination)
        return f"Flight Time: {self.time}"

    def set_time(self, time):
        self.time = f"{time}:00"

    def set_destination(self, des):
        self.destination = des

    def set_source(self, source):
        self.source = source


print("No. of Traveller =", Travel.count)
print("=======================")
t1 = Travel("Dhaka", "India")
print(t1.display_travel_info())
print("=======================")
t2 = Travel("Kuala Lampur", "Dhaka")
t2.set_time(23)
print(t2.display_travel_info())
print("=======================")
t3 = Travel("Dhaka", "New_Zealand")
t3.set_time(15)
t3.set_destination("Germany")
print(t3.display_travel_info())
print("=======================")
t4 = Travel("Dhaka", "India")
t4.set_time(9)
t4.set_source("Malaysia")
t4.set_destination("Canada")
print(t4.display_travel_info())
print("=======================")
print("No. of Traveller =", Travel.count)
```

```python
class NikeBangladesh:
    branches = []
    stock = {"Air Jordan": 0, "Cortez": 0, "Zoom Kobe": 0}
    sold = 0

    @classmethod
    def status(cls):
        print("Nike Bangladesh Status:")
        print("Branches Opened:", cls.branches)
        print("Currently Stocked")
        print(cls.stock)
        print("Sold:", cls.sold)

    def __init__(self, place):
        NikeBangladesh.branches.append(place)
        self.outlet = place
        self.stock = {"Air Jordan": 0, "Cortez": 0, "Zoom Kobe": 0}
        self.sold = 0

    def details(self):
        print(f"Nike {self.outlet} Status:")
        print("Products Currently Stocked")
        print(self.stock)
        print("Sold:", self.sold)

    def restockProducts(self, products):
        for key, value in products.items():
            self.stock[key] += value
            NikeBangladesh.stock[key] += value

    def productSold(self, products):
        for key, value in products.items():
            self.stock[key] -= value
            NikeBangladesh.stock[key] -= value
            NikeBangladesh.sold += value


print("xxxxxxxxxxxxxx1xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
dhaka = NikeBangladesh("Dhaka Banani")
chittagong = NikeBangladesh("Chittagong GEC")
print("xxxxxxxxxxxxxx2xxxxxxxxxxxxxxxx")
dhaka.details()
print("xxxxxxxxxxxxxx3xxxxxxxxxxxxxxxx")
chittagong.details()
print("xxxxxxxxxxxxxx4xxxxxxxxxxxxxxxx")
dhaka.restockProducts({"Air Jordan": 1200, "Cortez": 200, "Zoom Kobe": 200})
chittagong.restockProducts({"Air Jordan": 1000, "Cortez": 250, "Zoom Kobe": 100})
print("xxxxxxxxxxxxxx5xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
print("xxxxxxxxxxxxxx6xxxxxxxxxxxxxxxx")
dhaka.productSold({"Air Jordan": 760, "Cortez": 90})
chittagong.productSold({"Air Jordan": 520, "Zoom Kobe": 70})
print("xxxxxxxxxxxxxx7xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
```

```python
# Write your code here
class SultansDine:
    branches = 0
    sold = 0
    megaList = []

    @classmethod
    def details(cls):
        print("Total Number of branch(s):", cls.branches)
        print("Total Sell:", cls.sold, "Taka")

        for obj in cls.megaList:
            percentage = (obj.sells / cls.sold) * 100
            percentage = round(percentage, 2)

            print(f"Branch Name: {obj.outlet}, Branch Sell: {obj.sells}")
            print(f"Branch consists of total sell's: {percentage}%")

    def __init__(self, outlet):
        self.outlet = outlet
        self.sells = 0

        SultansDine.branches += 1
        SultansDine.megaList.append(self)

    def branchInformation(self):
        print("Branch Name:", self.outlet)
        print("Branch sell:", self.sells, "Taka")

    def sellQuantity(self, qty):
        if qty < 10:
            self.sells += qty * 300
            SultansDine.sold += qty * 300
        elif qty < 20:
            self.sells += qty * 350
            SultansDine.sold += qty * 350
        else:
            self.sells += qty * 400
            SultansDine.sold += qty * 400


SultansDine.details()
print("########################")
dhanmondi = SultansDine("Dhanmondi")
dhanmondi.sellQuantity(25)
dhanmondi.branchInformation()
print("-----------------------------------------")
SultansDine.details()
print("========================")
baily_road = SultansDine("Baily Road")
baily_road.sellQuantity(15)
baily_road.branchInformation()
print("-----------------------------------------")
SultansDine.details()
print("========================")
gulshan = SultansDine("Gulshan")
gulshan.sellQuantity(9)
gulshan.branchInformation()
print("-----------------------------------------")
SultansDine.details()
```
